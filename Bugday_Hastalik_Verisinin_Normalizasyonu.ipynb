```python
import os
from PIL import Image as img
import numpy as np
import pandas as pd
import random
import matplotlib.pyplot as plt
```

```python
dosya_yolu = r"C:\Users\gulic\Desktop\BİLGİ İŞLEM\Bugday\train"
```

```python
hastalik_adi = random.choice(os.listdir(dosya_yolu))
tam_yol = os.path.join(dosya_yolu,hastalik_adi)
resim = random.choice(os.listdir(tam_yol))
resim_yolu = os.path.join(tam_yol,resim)
orginal = img.open(resim_yolu)
```

```python
resim_img = np.array(orginal.convert("RGB").resize((244,244)), dtype=np.float32)/255
```

```python
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.title("Normalize edilmiş")
plt.imshow(resim_img)
plt.axis("off")

plt.subplot(1, 2, 2)
plt.title("Orjinal görüntü")
plt.imshow(orginal)
plt.axis("off")

plt.tight_layout()
plt.show()
```
