```python
import os
from PIL import Image as img
import numpy as np
import pandas as pd
import random
import matplotlib.pyplot as plt
```

```python
dosya_yolu = r"C:\Users\gulic\Desktop\BİLGİ İŞLEM\Bugday\train"
```

```python
hastalik_adi = random.choice(os.listdir(dosya_yolu))
tam_yol = os.path.join(dosya_yolu,hastalik_adi)
resim = random.choice(os.listdir(tam_yol))
resim_yolu = os.path.join(tam_yol,resim)
resim_img = np.array(img.open(resim_yolu).convert("RGB").resize((244,244)), dtype=np.float32)/255
arr_norm_uint8 = (resim_img * 255).astype(np.uint8) 
img_norm = Image.fromarray(arr_norm_uint8)
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.title("Orijinal (Resize Edilmiş)")
plt.imshow(resim_img)
plt.axis("off")

plt.subplot(1, 2, 2)
plt.title("Normalize Edilmiş Görüntü")
plt.imshow(img_norm)
plt.axis("off")

plt.tight_layout()
plt.show()
```
