```python
import os
import random
import shutil
```

```python
tam_yol = r"C:\Users\gulic\Desktop\BİLGİ İŞLEM\Bugday\deneme"
```

```python
siniflar = os.listdir(tam_yol)
siniflar
```

```python
train_N = 0.7
val_N = 0.15
test_N = 0.15
for sinif in siniflar:
    resim_yolu = os.path.join(tam_yol,sinif)
    train_yol = os.path.join(tam_yol,"train",sinif)
    test_yol = os.path.join(tam_yol,"test",sinif) 
    val_yol = os.path.join(tam_yol,"val",sinif) 
    for kisim in (train_yol,test_yol, val_yol):
        os.makedirs(kisim,exist_ok = True)
    resimler = [ a for a in os.listdir(resim_yolu) if a.lower().endswith((".jpg",".png", ".jpeg"))]
    random.shuffle(resimler)
    train_ = int(train_N*len(resimler))
    test_ = int(test_N*len(resimler))
    val_ = int(val_N*len(resimler))
    train_resimler = resimler[:train_]
    test_resimler = resimler[train_:train_+test_]
    val_resimler = resimler[train_+test_:]
    def kopyala(kaynak,hedef):
        for resim in kaynak:
            shutil.copy(os.path.join(resim_yolu,resim), os.path.join(hedef,resim))
    kopyala(train_resimler,train_yol)
    kopyala(test_resimler,test_yol)
    kopyala(val_resimler,val_yol)
    print(f"{sinif} :train: {train_} test: {test_} val: {val_}")
```
